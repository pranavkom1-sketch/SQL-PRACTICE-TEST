7. Percentile is a statistical measure that indicates the value below which a certain percentage of the data falls.
 
   25th percentile (P25): 25% of the data is below this value.

   50th percentile (P50): 50% of the data is below this value (also the Median).

   75th percentile (P75): 75% of the data is below this value.
 
   Percentiles are commonly used in exam scores, employee salaries.

8. Quartiles divide a data set into four equal parts. After it has been arranged in ascending order.

   First Quartile (Q1) – 25th Percentile - This is the median of the lower half of the data (below the overall median).

   Second Quartile (Q2) – 50th Percentile (Median) - middle value of the data set when arranged in ascending order. It divides the data into two equal halves.

   Third Quartile (Q3) – 75th Percentile - This is the median of the upper half of the data (above the overall median).


   
   Interquartile Range (IQR) - it is the difference between the third quartile and the first quartile.

   IQR=Q3- Q1
   Lower Boundary=𝑄1−1.5×IQR
   Upper Boundary=𝑄3+1.5×IQR
   These boundaries are used to detect outliers. Any data point below the lower boundary or above the upper boundary would be considered an outlier.

10. A Box Plot (also known as a Box-and-Whisker Plot) is a graphical representation of a data set that displays its minimum, first quartile (Q1),
    median (Q2), third quartile (Q3), and maximum values.
    
    It also shows potential outliers.
    
    Box: Represents the interquartile range (IQR) from Q1 to Q3.
    
    Whiskers: The lines extending from the box, indicating the range from the minimum to Q1 and from Q3 to the maximum.
    
    Median: The line inside the box, showing the middle value (Q2).
    
    Outliers: Any data points that fall outside the lower boundary and upper boundary (calculated as 1.5*IQR from Q1 and Q3). These are typically marked with dots or asterisks.

11. Skewness is a measure of the asymmetry in the distribution of a dataset.
    
    It describes whether the data is symmetrically distributed, or if it is skewed towards one side (left or right).

    a) Symmetrical Distribution (Skewness = 0) - The data is evenly distributed around the mean.

       Mean = Median = Mode

    b) Positive Skewness (Right-Skewed) - The tail of the distribution extends more towards the right (higher values).

       Mean > Median > Mode.
    
    c) Negative Skewness (Left-Skewed) - The tail of the distribution extends more towards the left (lower values).
    
       Mean < Median < Mode.

15. Covariance measures how two random variables change together.
    
    Correlation is a standardized measure of the strength and direction of the linear relationship between two variables.

    a)Positive Covariance: When one variable increases, the other variable tends to increase as well and vice versa.

    b)Negative Covariance: When one variable increases, the other variable tends to decrease.
    
    c)Zero Covariance: There is no linear relationship between the two variables; they move independently of each other.

    d)Positive Correlation (close to +1): As one variable increases, the other variable also tends to increase.
    
    e)Negative Correlation (close to -1): As one variable increases, the other variable tends to decrease.
    
    f)Zero Correlation: There is no linear relationship between the variables.
